```python
#Using the SQL workbench on Udacity I was able to pull the data using the following two SQL queries. 
"""
SELECT * FROM city_data WHERE city LIKE '%Houston%'
SELECT * FROM global_data
"""

#Loading data from csv file, for both city and global data.
city_data = pd.read_csv('city_data.csv')
#Renaming the global data colunms for clearity. 
global_cols = ['year', 'global_avg_temp']
global_data = pd.read_csv('global_data.csv', names=global_cols, skiprows=1)
```

##**Considerations and Observations**

When comparing the global with the average tempurature in Houston, TX we can see that the tempurature does vary using a window of 1 for the moving average both for the global and the city date, we can clearly see the spike for this plot. Using the line plot we can we see when we move from 1 year to 5, we notice that the moving average line starts to get smoother, as well as both the 7 and 10 year marks too. 


We can see that the min and max for both the world and for Houston, Tx does vary. While the avg are different we can see from our line plots that we are moving in the same direction over time. We see that from the min and max for the global and Houston never passed over 4.05 for that threshold.


```python
#Moving averages for 1 year
city_data['houston_ma'] = city_data['avg_temp'].rolling(window=1).mean()
global_data['global_ma'] = global_data['global_avg_temp'].rolling(window=1).mean()

#storing temperature and years for colunm values for plotting
#Global
global_ma = global_data['global_avg_temp']
year_global = global_data['year']
#City
houston_ma = city_data['houston_ma']
year_houston = city_data['year']

fig = plt.figure(figsize=(20, 17))
plt.subplot(4,1,1)
plt.grid()
plt.plot(year_global,global_ma, label='Global Temp Over Time')
plt.plot(year_houston,houston_ma, label='Houston Temp Over Time')
plt.legend(loc=2)
plt.ylim(5,30)
plt.xlim(1750,2020)
plt.title('Average Tempature Over Time - Moving Average for 1Y',fontsize=22)
plt.xlabel('Years',fontsize=17)
plt.ylabel('Temperature in Â°C',fontsize=15)
```

```python
#Moving averages for 5 years
city_data['houston_ma'] = city_data['avg_temp'].rolling(window=5).mean()
global_data['global_ma'] = global_data['global_avg_temp'].rolling(window=5).mean()
fig = plt.figure(figsize=(20, 17))
plt.subplot(4,1,1)
plt.grid()
plt.plot(year_global,global_ma, label='Global Temp Over Time')
plt.plot(year_houston,houston_ma, label='Houston Temp Over Time')
plt.legend(loc=2)
plt.ylim(5,30)
plt.xlim(1750,2020)
plt.title('Average Tempature Over Time - Moving Average for 5Y',fontsize=22)
plt.xlabel('Years',fontsize=17)
plt.ylabel('Temperature in Â°C',fontsize=15)
```

```python
#Moving averages for 7 years
city_data['houston_ma'] = city_data['avg_temp'].rolling(window=7).mean()
global_data['global_ma'] = global_data['global_avg_temp'].rolling(window=7).mean()
fig = plt.figure(figsize=(20, 17))
plt.subplot(4,1,1)
plt.grid()
plt.plot(year_global,global_ma, label='Global Temp Over Time')
plt.plot(year_houston,houston_ma, label='Houston Temp Over Time')
plt.legend(loc=2)
plt.ylim(5,30)
plt.xlim(1750,2020)
plt.title('Average Tempature Over Time - Moving Average for 7Y',fontsize=22)
plt.xlabel('Years',fontsize=17)
plt.ylabel('Temperature in Â°C',fontsize=15)
```

```python
#Moving averages for 10 years
city_data['houston_ma'] = city_data['avg_temp'].rolling(window=10).mean()
global_data['global_ma'] = global_data['global_avg_temp'].rolling(window=10).mean()
fig = plt.figure(figsize=(20, 17))
plt.subplot(4,1,1)
plt.grid()
plt.plot(year_global,global_ma, label='Global Temp Over Time')
plt.plot(year_houston,houston_ma, label='Houston Temp Over Time')
plt.legend(loc=2)
plt.ylim(5,30)
plt.xlim(1750,2020)
plt.title('Average Tempature Over Time - Moving Average for 10Y',fontsize=22)
plt.xlabel('Years',fontsize=17)
plt.ylabel('Temperature in Â°C',fontsize=15)
```

```python
#Global
min_temp_global = global_data['global_avg_temp'].min()
max_temp_global = global_data['global_avg_temp'].max()
avg_temp_global = global_data['global_avg_temp'].mean()

#Houston, Texas
min_temp_city = city_data['avg_temp'].min()
max_temp_city = city_data['avg_temp'].max()
avg_temp_city = city_data['avg_temp'].mean()
```
